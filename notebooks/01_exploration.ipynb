```python
df = df.drop_duplicates()
df.shape
```

```python
df = df.drop(columns=['Unnamed: 0'])
df.columns.tolist()
```

```python
text_month_cols = ['StartDate', 'ExitDate', 'Training Date']
for col in text_month_cols:
    df[col] = pd.to_datetime(df[col], format='%d-%b-%y', errors='coerce')

numeric_date_cols = ['DOB', 'Survey Date']
for col in numeric_date_cols:
    df[col] = pd.to_datetime(df[col], format='%d-%m-%Y', errors='coerce')

df.dtypes
```

```python
df[text_month_cols + numeric_date_cols].isnull().sum()
```

```python
df['Attrition'] = df['ExitDate'].notna()
df['Attrition'].value_counts()
```

```python
df['GenderCode'].value_counts()
```

```python
df['RaceDesc'].value_counts()
```

```python
# Supervisor sloupec ponechán záměrně - budeme zkoumat, zda attrition
# souvisí s konkrétními manažery/týmy. Ostatní PII sloupce odstraněny.
pii_cols = ['FirstName', 'LastName', 'ADEmail']
df = df.drop(columns=pii_cols)
df.shape
```

```python
df.to_csv('../data/processed/HR_Dataset_Clean.csv', index=False)
```
